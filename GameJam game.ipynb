```python
#calculate text placement

def text_coords(text, npc):
    x = npc.x + GUY_WIDTH//2 - text.get_width()//2
    y = npc.y - HOVER - UNDER*2
    
    if x < WALLTHICK + 20:
        x = WALLTHICK + 20
    if y < WALLTHICK + 20:
        y = WALLTHICK + 20
        
    if x + text.get_width() > SCRN_WIDTH - WALLTHICK - 20:
        x = SCRN_WIDTH - WALLTHICK - 20 - text.get_width()
    
        
    return [x,y]
    
```

```python
#draw space prompt in given colour

def draw_space(colour):
    press_text = SMALLFONT.render("PRESS", 1, colour)
    space_text = BIGFONT.render("SPACE", 1, colour)
    WIN.blit(press_text, (SCRN_WIDTH//2 -23, BWALL.y+WALLTHICK + 25))
    WIN.blit(space_text, (SCRN_WIDTH//2 -65, BWALL.y+WALLTHICK + 50))
    
```

```python
#draw correctly sized text box bubble

def draw_bubble(t1, x1, y1, t2):
    
    if t2 != False and t2.get_width() > t1.get_width():
        width = t2.get_width()
        
    else:
        width = t1.get_width()
    
    cap = pg.Rect(x1-ALNTXT-BORWID, y1-ALNTXT-BORWID, BORWID*2 + ALNTXT*2 + width, BORDER_HEIGHT)
    pg.draw.rect(WIN, YELLOW, cap)
    pg.draw.rect(WIN, GREY, pg.Rect(x1-ALNTXT, y1-ALNTXT, ALNTXT*2 + width, BOX_HEIGHT))
    
    return cap
    
```

```python
#see if we can start convo with given npc

def check_in_zone(cm, npc):
    if npc.x-GUY_WIDTH < cm.x < npc.x+GUY_WIDTH+10 and npc.y < cm.y < npc.y+GUY_HEIGHT*2:
        return True
    else:
        return False
```

```python
#see if carpetman is in a certain area

def check_in_area(cm, box):
    if box.x <= cm.x <= box.x + box.width - GUY_WIDTH and box.y - 20 <= cm.y <= box.y + box.height - GUY_HEIGHT:
        return True
    else:
        return False
```

```python
#room 3 collision checkers

def check_collide_up3(cm, cap):
    if cm.y - SPD < TWALL3.y + TWALL3.height:
        return False
    if cm.x < LWALL3.width:
        if cm.y - SPD < LWALL3.height:
            return False
    if JUT.x - GUY_WIDTH <= cm.x <= JUT.x + JUT.width and cm.y >= JUT.y + JUT.height:
        if cm.y - SPD < JUT.y + JUT.height:
            return False
    if cm.x > JUT.x + JUT.width and cm.y >= BWALL3.y + BWALL3.height:
        if cm.y - SPD < BWALL3.y + BWALL3.height:
            return False
    return True

def check_collide_down3(cm, cap):
    if cm.y + SPD + GUY_HEIGHT > BWALL.y:
        return False
    if cm.y <  BWALL3.y +4 and cm.x >= BWALL3.x - GUY_WIDTH:
        
        if cm.y + SPD + GUY_HEIGHT > BWALL3.y:
            if cap.x < cm.x < cap.x +cap.width - GUY_WIDTH and cap.y < cm.y + GUY_HEIGHT +SPD < cap.y + cap.height:
                cap.x = cap.x
            else:
                return False
    if cm.y + GUY_HEIGHT <= cap.y +cap.height and cm.x + GUY_WIDTH >= JUT.x:
        if cm.y + SPD + GUY_HEIGHT > cap.y +cap.height:
            return False
    return True

def check_collide_left3(cm, cap):
    if cm.x - SPD < LWALL.x + WALLTHICK:
        return False
    if cm.x >= LWALL3.width and cm.y <= LWALL3.height:
        if cm.x - SPD < LWALL3.width:
            return False
    if cm.x >= JUT.x + JUT.width and  JUT.y <= cm.y <= JUT.y + JUT.height:
        if cm.x - SPD < JUT.x + JUT.width:
            if cap.x < cm.x -SPD < cap.x +cap.width - GUY_WIDTH and cap.y < cm.y + GUY_HEIGHT < cap.y + cap.height:
                cap.x = cap.x
            else:
                return False
    
    return True

def check_collide_right3(cm, cap):
    if cm.x + SPD + GUY_WIDTH > RWALL.x:
        return False
    if JUT.y - GUY_HEIGHT < cm.y < JUT.y + JUT.height and cm.x < JUT.x:
        if cm.x + SPD + GUY_WIDTH > JUT.x:
            if cap.x < cm.x +SPD < cap.x +cap.width - GUY_WIDTH and cap.y < cm.y + GUY_HEIGHT < cap.y + cap.height:
                cap.x = cap.x
            else:
                return False
    if cm.x +SPD > cap.x +cap.width - GUY_WIDTH and cap.y < cm.y + GUY_HEIGHT < cap.y + cap.height:
        return False
        
    return True
```

```python
def carpetman_move3(keys_pressed, cm, cap):
    if (keys_pressed[pg.K_w] or keys_pressed[pg.K_UP]) and check_collide_up3(cm, cap):
        cm.y -= SPD
    if (keys_pressed[pg.K_s] or keys_pressed[pg.K_DOWN]) and check_collide_down3(cm, cap):
        cm.y += SPD
    if (keys_pressed[pg.K_a] or keys_pressed[pg.K_LEFT]) and check_collide_left3(cm, cap):
        cm.x -= SPD
    if (keys_pressed[pg.K_d] or keys_pressed[pg.K_RIGHT]) and check_collide_right3(cm, cap):
        cm.x += SPD
```

```python
def calculate(cm, npc):
    theta = np.arctan2(cm.y - npc.y, cm.x - npc.x)
    npc.y += round(BOSSSPD*np.sin(theta))
    npc.x += round(BOSSSPD*np.cos(theta))
    
    if npc.x < WALLTHICK or npc.x > RWALL.x or npc.y < WALLTHICK or npc.y > BWALL.y:
        return True
    
    return False
    
```

```python
def damage_cm(cm, npc, tc):
    tolerance = 20
    if tc[0] - GUY_WIDTH < cm.x < tc[0] + 280 and tc[1] - GUY_WIDTH < cm.y < tc[1] + 90:
        BLIP2.play()
        return True
    if npc.x - GUY_WIDTH + tolerance < cm.x < npc.x + GUY_WIDTH - tolerance and npc.y - GUY_WIDTH + tolerance < cm.y < npc.y + GUY_WIDTH - tolerance:
        BLIP2.play()
        return True
    return False
```

```python
#room 5 collision checkers

def check_collide_up5(cm, cap):
    
    #wall
    if cm.y - SPD < TWALL.y + WALLTHICK:
        return False
    
    #if below ravine
    if cm.x <= BRAV.width and cm.y >= BRAV.y + BRAV.height:
        #if would go in ravine
        if cm.y - SPD < BRAV.y + BRAV.height:
            return False
    
    #if under/on RRAV
    if RRAV.x - GUY_WIDTH < cm.x < RRAV.x + RRAV.width:
        # if go above caption
        if cm.y - SPD < cap.y:
            return False
        
    return True

def check_collide_down5(cm, cap):
    #wall
    if cm.y + SPD + GUY_HEIGHT > BWALL.y:
        return False
    
    #if above BRAV
    if cm.x < BRAV.width and cm.y < BRAV.y - GUY_HEIGHT:
        #if go into rav
        if cm.y + SPD + GUY_HEIGHT > BRAV.y:
            return False
        
    if RRAV.x - GUY_WIDTH < cm.x < RRAV.x + RRAV.width and cm.y < BRAV.y:
        if cm.y + SPD + GUY_HEIGHT > cap.y + cap.height:
            return False
    
    return True


def check_collide_left5(cm, cap):
    #wall
    if cm.x - SPD < LWALL.x + WALLTHICK:
        return False
    
    #to right of ravine
    if cm.x > RRAV.x + RRAV.width and cm.y < RRAV.y + RRAV.height:
        #if would go in ravine
        if cm.x - SPD < RRAV.x + RRAV.width:
            #if alighned with caption and moving will not leave the caption
            if cap.y + cap.height - GUY_HEIGHT > cm.y > cap.y and cm.x - SPD > cap.x:
                return True
            else:
                return False
        
    return True

def check_collide_right5(cm, cap):
    #wall
    if cm.x + SPD + GUY_WIDTH > RWALL.x:
        return False
    
    #if to left of rav
    if cm.x <= RRAV.x and cm.y <= RRAV.y + RRAV.height:
        #if will go in ravine
        if cm.x + SPD + GUY_WIDTH > RRAV.x:
            #if will land on caption
            if cap.y + cap.height - GUY_HEIGHT > cm.y > cap.y and cm.x + SPD > cap.x:
                return True
            else:
                return False
    
    return True
```

```python
def carpetman_move5(keys_pressed, cm, cap):
    if (keys_pressed[pg.K_w] or keys_pressed[pg.K_UP]) and check_collide_up5(cm, cap):
        cm.y -= SPD
    if (keys_pressed[pg.K_s] or keys_pressed[pg.K_DOWN]) and check_collide_down5(cm, cap):
        cm.y += SPD
    if (keys_pressed[pg.K_a] or keys_pressed[pg.K_LEFT]) and check_collide_left5(cm, cap):
        cm.x -= SPD
    if (keys_pressed[pg.K_d] or keys_pressed[pg.K_RIGHT]) and check_collide_right5(cm, cap):
        cm.x += SPD
```

```python
def room7(health):
    cm = pg.Rect(SCRN_WIDTH/2 -25, 600, GUY_WIDTH, GUY_HEIGHT)
    tnpc = pg.Rect(SCRN_WIDTH/2 -25, 200, GUY_WIDTH, GUY_HEIGHT)
    
    space = 0
    talk_step = 0
    
    clock = pg.time.Clock()
    in_zone = False
    run = True
    endin = 40
    notpressed = True
    while run:
        clock.tick(FPS)
        for event in pg.event.get():
            if event.type == pg.QUIT:
                return False
            
            if event.type == pg.KEYDOWN:
                if event.key == pg.K_SPACE and (in_zone or talk_step > 0):
                    talk_step = (talk_step +1)%(TUTLINES+1)
                    BLIP.play()
                if event.key == pg.K_x:
                    print(cm.x)
                if event.key == pg.K_c:
                    print(cm.y)
                    
            if event.type == MOVE_ROOM2:
                run = False
                
            
        draw_room7(cm, tnpc, space, talk_step, health, notpressed)
        
        keys_pressed = pg.key.get_pressed()
        carpetman_move(keys_pressed, cm)
        
        in_zone = check_in_zone(cm, tnpc)
        if in_zone:
            if talk_step > 0:
                space = 2
            else:
                space = 1
        else:
            if talk_step > 0:
                space = 2
            else:
                space = 0
                
        if space == 2 and notpressed:
            notpressed = False
            
        if not notpressed:
            endin -= 1
            if endin <=0:
                return False
            
    return True
```
